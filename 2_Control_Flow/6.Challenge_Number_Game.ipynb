# Number Guessing Game
Do you want to play a game? The computer wants you to guess which number it is thinking of from 1 to 10 (including 1 and 10).

The script should:
- Generate a random number between 1 and 10 (inclusive)
- Prompt the user to enter a guess
  - **Hint**: Convert the user's input from string to integer
    ```[int](Read-Host -Prompt "`nGuess a number")```
- Tell the user if their guess is higher, lower, or matches the random number

The script output should look something like this:
```
Let's play a game! I'm thinking of a number from 1 to 10. Can you guess which one?

Guess a number: 5
Your guess is too low, try again.

Guess a number: 8
Your guess too high, try again.

Guess a number: 7
Your guess too high, try again.

Guess a number: 6
You guessed correctly!
Game over!
```
